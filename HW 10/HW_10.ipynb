```python
import pandas as pd
import random
lst = ['robot'] * 10
lst += ['human'] * 10
random.shuffle(lst)
data = pd.DataFrame({'whoAmI':lst})
data
```

```python
data_test = data
pd.get_dummies(data)
```

```python
data.loc[data['whoAmI'] == 'human', 'whoAmI_human'] = 1
data.loc[data['whoAmI_human'] != 1, 'whoAmI_human'] = 0

data.loc[data['whoAmI'] == 'robot', 'whoAmI_robot'] = 1
data.loc[data['whoAmI_robot'] != 1, 'whoAmI_robot'] = 0

data.drop('whoAmI', axis=1)
```
